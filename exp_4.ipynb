```python
import gdown
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img

# Step 1: Download from Google Drive
file_id = "1nBob8sXuCCxCp4I2-2hCTmctAkgNEven"
gdown.download(f"https://drive.google.com/file/d/1nBob8sXuCCxCp4I2-2hCTmctAkgNEven/view?usp=drive_link/uc?id={file_id}", "sheep.jpg", quiet=False)

# Step 2: Load and preprocess image
img_path = "/content/butterfly.jpg"
image = load_img(img_path, target_size=(224, 224))
image_array = img_to_array(image)
image_array = np.expand_dims(image_array, axis=0)

# Step 3: Show original image
plt.figure(figsize=(5, 5))
plt.imshow(image_array[0].astype("uint8"))
plt.title("Original Image")
plt.axis("off")
plt.show()

# Step 4: Define ImageDataGenerator with augmentations
datagen = ImageDataGenerator(
    rotation_range=40,
    zoom_range=0.2,
    brightness_range=[0.5, 1.5],
    shear_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

# Step 5: Generate and display augmented images
plt.figure(figsize=(10, 10))
i = 0
for batch in datagen.flow(image_array, batch_size=1):
    plt.subplot(3, 3, i + 1)
    plt.imshow(batch[0].astype("uint8"))
    plt.axis("off")
    i += 1
    if i >= 9:
        break
plt.suptitle("Augmented Images")
plt.show()

```
